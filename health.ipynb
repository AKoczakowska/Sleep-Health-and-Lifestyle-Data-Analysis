W trakcie analizy poziomu stresu w zaleÅ¼noÅ›ci od wykonywanego zawodu, napotkaÅ‚am na kilka grup zawodowych z bardzo maÅ‚Ä… liczbÄ… reprezentantÃ³w w danych. Dla zapewnienia lepszej wiarygodnoÅ›ci statystycznej, postanowiÅ‚am odrzuciÄ‡ zawody, w ktÃ³rych liczba osÃ³b wynosiÅ‚a mniej niÅ¼ 5.

#### Powody odrzucenia:
- **Statystyczna niereprezentatywnoÅ›Ä‡**: W przypadku zawodÃ³w z bardzo maÅ‚Ä… liczbÄ… obserwacji (n < 5), Å›redni poziom stresu moÅ¼e byÄ‡ silnie zaburzony przez jednostkowe, skrajne wartoÅ›ci, co prowadziÅ‚oby do nieprawidÅ‚owych wnioskÃ³w.
- **WiarygodnoÅ›Ä‡ analizy**: Praca z wiÄ™kszymi grupami pozwala na bardziej stabilne i reprezentatywne porÃ³wnania pomiÄ™dzy rÃ³Å¼nymi zawodami.
- **Ograniczenia prÃ³by**: W projekcie uwzglÄ™dniÅ‚am tylko te grupy zawodowe, ktÃ³re majÄ… dostatecznie duÅ¼Ä… liczbÄ™ reprezentantÃ³w, aby moÅ¼liwe byÅ‚o wyciÄ…ganie wiarygodnych statystycznie wnioskÃ³w.

#### Zawody odrzucone:
Odrzucone zawody ze wzglÄ™du na zbyt maÅ‚Ä… liczbÄ™ danych to:
- Manager (1 osoba)
- Sales Representative (2 osoby)
- Software Engineer (4 osoby)
- Scientist (4 osoby)

Grupy te zostaÅ‚y wyÅ‚Ä…czone z analizy poziomu stresu, aby uniknÄ…Ä‡ bÅ‚Ä™dnych wnioskÃ³w na podstawie niewystarczajÄ…cych danych. Analiza poziomu stresu dla pozostaÅ‚ych grup zawodowych zostaÅ‚a przeprowadzona z zachowaniem peÅ‚nej rzetelnoÅ›ci.


```python
valid_occupations = OccupationCount[OccupationCount >=5].index
filtered_occupations = data[data['Occupation'].isin(valid_occupations)]
filtered_occupations
filtered_occupations_mean = filtered_occupations.groupby(['Occupation'])['Stress Level'].mean()
filtered_occupations_mean = pd.DataFrame(filtered_occupations_mean)
filtered_occupations_mean
```

```python
filtered_occupations_mean.sort_values(by='Stress Level', ascending=False, inplace=True)
max_index = np.argmax(filtered_occupations_mean['Stress Level'])
colors = ['orange' if i == max_index else 'lightblue' for i in range(len(filtered_occupations_mean))]
filtered_occupations_mean['Color'] = colors

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=filtered_occupations_mean.index, y='Stress Level', data=filtered_occupations_mean, palette=colors)
plt.title('Average stress levels for occupations')
plt.xlabel('Occupation')
plt.ylabel('Average stress levels')
for i, value in enumerate(filtered_occupations_mean['Stress Level']):
    plt.text(i, value, str(round(value, 2)), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.ylim(0, 8)
plt.tight_layout()
plt.show()
```
