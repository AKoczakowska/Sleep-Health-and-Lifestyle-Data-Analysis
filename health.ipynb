### Odrzucenie zawodów o małej liczbie reprezentantów

W trakcie analizy poziomu stresu w zależności od wykonywanego zawodu, napotkałam na kilka grup zawodowych z bardzo małą liczbą reprezentantów w danych. Dla zapewnienia lepszej wiarygodności statystycznej, postanowiłam odrzucić zawody, w których liczba osób wynosiła mniej niż 5.

#### Powody odrzucenia:
- **Statystyczna niereprezentatywność**: W przypadku zawodów z bardzo małą liczbą obserwacji (n < 5), średni poziom stresu może być silnie zaburzony przez jednostkowe, skrajne wartości, co prowadziłoby do nieprawidłowych wniosków.
- **Wiarygodność analizy**: Praca z większymi grupami pozwala na bardziej stabilne i reprezentatywne porównania pomiędzy różnymi zawodami.
- **Ograniczenia próby**: W projekcie uwzględniłam tylko te grupy zawodowe, które mają dostatecznie dużą liczbę reprezentantów, aby możliwe było wyciąganie wiarygodnych statystycznie wniosków.

#### Zawody odrzucone:
Odrzucone zawody ze względu na zbyt małą liczbę danych to:
- Manager (1 osoba)
- Sales Representative (2 osoby)
- Software Engineer (4 osoby)
- Scientist (4 osoby)

Grupy te zostały wyłączone z analizy poziomu stresu, aby uniknąć błędnych wniosków na podstawie niewystarczających danych. Analiza poziomu stresu dla pozostałych grup zawodowych została przeprowadzona z zachowaniem pełnej rzetelności.


```python
valid_occupations = OccupationCount[OccupationCount >=5].index
filtered_occupations = data[data['Occupation'].isin(valid_occupations)]
filtered_occupations
filtered_occupations_mean = filtered_occupations.groupby(['Occupation'])['Stress Level'].mean()
filtered_occupations_mean = pd.DataFrame(filtered_occupations_mean)
filtered_occupations_mean
```

```python
filtered_occupations_mean.sort_values(by='Stress Level', ascending=False, inplace=True)
max_index = np.argmax(filtered_occupations_mean['Stress Level'])
colors = ['orange' if i == max_index else 'lightblue' for i in range(len(filtered_occupations_mean))]
filtered_occupations_mean['Color'] = colors

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=filtered_occupations_mean.index, y='Stress Level', data=filtered_occupations_mean, palette=colors)
plt.title('Average stress levels for occupations')
plt.xlabel('Occupation')
plt.ylabel('Average stress levels')
for i, value in enumerate(filtered_occupations_mean['Stress Level']):
    plt.text(i, value, str(round(value, 2)), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.ylim(0, 8)
plt.tight_layout()
plt.show()
```
