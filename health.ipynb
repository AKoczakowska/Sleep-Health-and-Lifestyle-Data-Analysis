### Exclusion of Occupations with a Small Number of Representatives

During the analysis of stress levels depending on the occupation, I encountered several occupational groups with a very small number of representatives in the data. To ensure better statistical reliability, I decided to exclude occupations where the number of individuals was less than 5.

#### Reasons for Exclusion:
- **Statistical Unrepresentativeness**: For occupations with a very small number of observations (n < 5), the average stress level might be heavily distorted by individual, extreme values, leading to incorrect conclusions.
- **Analysis Reliability**: Working with larger groups allows for more stable and representative comparisons between different occupations.
- **Sample Limitations**: The study included only those occupational groups with a sufficiently large number of representatives to draw statistically reliable conclusions.

#### Excluded Occupations:
Occupations excluded due to insufficient data include:
- Manager (1 person)
- Sales Representative (2 people)
- Software Engineer (4 people)
- Scientist (4 people)
- These groups were excluded from the stress level analysis to avoid erroneous conclusions based on insufficient data. The analysis of stress levels for the remaining occupational groups was conducted with full reliability.


```python
valid_occupations = OccupationCount[OccupationCount >=5].index
filtered_occupations = data[data['Occupation'].isin(valid_occupations)]
filtered_occupations
filtered_occupations_mean = filtered_occupations.groupby(['Occupation'])['Stress Level'].mean()
filtered_occupations_mean = pd.DataFrame(filtered_occupations_mean)
filtered_occupations_mean
```

```python
filtered_occupations_mean.sort_values(by='Stress Level', ascending=False, inplace=True)
max_index = np.argmax(filtered_occupations_mean['Stress Level'])
colors = ['orange' if i == max_index else 'lightblue' for i in range(len(filtered_occupations_mean))]
filtered_occupations_mean['Color'] = colors

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=filtered_occupations_mean.index, y='Stress Level', data=filtered_occupations_mean, palette=colors)
plt.title('Average stress levels for occupations')
plt.xlabel('Occupation')
plt.ylabel('Average stress levels')
for i, value in enumerate(filtered_occupations_mean['Stress Level']):
    plt.text(i, value, str(round(value, 2)), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.ylim(0, 8)
plt.tight_layout()
plt.show()
```

```python
correlation = data['Quality of Sleep'].corr(data['Stress Level'])
print("Korelacja Pearsona:", correlation)
```

A Pearson correlation result of -0.899 indicates a strong negative correlation between sleep quality and stress levels. This means that as sleep quality improves, stress levels tend to decrease, and as sleep quality worsens, stress levels increase. This value suggests a clear linear relationship, where one phenomenon changes in the opposite direction to the other.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
plot = sns.regplot(
    x='Quality of Sleep', 
    y='Stress Level', 
    data=data,
    scatter_kws={'color':'steelblue', 's':100, 'edgecolor':'black'},
    line_kws={'color':'darkred', 'linewidth':2},
    ci=None  
)
plt.title('Relationship between sleep quality and stress levels', fontsize=16, weight='bold')
plt.xlabel('Quality of Sleep', fontsize=14)
plt.ylabel('Stress Level', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
